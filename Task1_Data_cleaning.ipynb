## Task 1: Data Cleaning and Preprocessing

```python
# Pandas is used for data manipulation and analysis
import pandas as pd
```

```python
# Numpy is used for numerical operation
import numpy as np
```

```python
#Reading customer churn dataset
df=pd.read_csv("Telco_Customer_Churn_Dataset .csv")
```

```python
# Display first five row of dataset
df.head()
```

```python
# Check info of dataset such as dat type and non-null values
df.info()
```

#### We can see there is no null values in any column so we don't need to fill the values with mean or mode

```python
# Check number of rows & col
df.shape
```

```python
df.drop_duplicates(inplace=True)
```

```python
# TotalCharges should be numeric but it is stored as object
# converting it into nueric datatype
df['TotalCharges']= pd.to_numeric(df['TotalCharges'],errors='coerce')
```

```python
df['TotalCharges']
```

```python
# CostumerID is an identifier and doesn't help in prediction 
# So we should remove this before prediction
df.drop('customerID',axis=1,inplace=True)
```

```python
# Converting categorial values into boolean or numerical formateusing one hot encoding
# drop_first=True avoid dummy variable trap
df_encoded=pd.get_dummies(df,drop_first=True)
```

```python
df_encoded.head()
```

#### ML algorithms cannot understand text like yes/no or Male/Female
#### Encoding convert them into numeric or boolean form

```python
df_encoded.shape
```

```python
df_encoded.to_csv('cleaned_customer_churn.csv',index=False)
```
